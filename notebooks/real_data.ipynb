## Understanding the scores inside the manifold

```python
tau95 = 2.268063545227051
tau99 = 3.7468080520629883
```

```python
(score["conduct_score_centered"] > tau95).mean()
```

```python
((score["conduct_score_centered"] > tau95) & (~score["recon_outside95"])).mean()
```

```python
((score["conduct_score_centered"] > tau95) & (score["recon_outside95"])).mean()
```

```python
pd.crosstab(
    score["conduct_score_centered"] > tau95,
    score["recon_outside95"])
```

```python
score["group"] = "Competitive / In manifold"

score.loc[
    (score["conduct_score_centered"] > tau95) &
    (~score["recon_outside95"]),
    "group"
] = "Cartel-like / In manifold"

score.loc[
    (score["conduct_score_centered"] <= tau95) &
    (score["recon_outside95"]),
    "group"
] = "Competitive / Out of manifold"

score.loc[
    (score["conduct_score_centered"] > tau95) &
    (score["recon_outside95"]),
    "group"
] = "Cartel-like/ Out of manifold"
```

```python
recon_tau95 = 0.268067
recon_tau99 = 0.610001

fig = px.scatter(
    score,
    x="recon_error",
    y="conduct_score_centered",
    color="group",
    opacity=0.5,
    height=600,
)

# log scale fixes the extreme skew
fig.update_xaxes(type="log")

# threshold lines
fig.add_vline(x=recon_tau95, line_dash="dash")
fig.add_hline(y=tau95, line_dash="dash")

fig.update_layout(
    title="Conduct Score vs Reconstruction Error",
    xaxis_title="Reconstruction Error (log scale)",
    yaxis_title="Centered Conduct Score"
)

fig.add_annotation(x=-2.5, y=15, text="Cartel-like<br>In manifold", showarrow=False)
fig.add_annotation(x=2, y=15, text="Cartel-like<br>Out of manifold", showarrow=False)
fig.add_annotation(x=-2.5, y=-25, text="Competitive", showarrow=False)
fig.add_annotation(x=2, y=-25, text="Anomalous", showarrow=False)

fig.show()
```

## Market level metrics

```python
market_metrics = compute_market_metrics_real(
    score,
    tau95=tau95,
    tau99=tau99,
    recon_tau95=recon_tau95,
    recon_tau99=recon_tau99,
    market_col="Name",
    time_col="Window"   
)
```

```python
market_metrics.head()
```

```python
market_metrics.columns
```

```python
def zscore(x):
    x = pd.Series(x)
    s = x.std(ddof=0)
    if s == 0 or pd.isna(s):
        return pd.Series(np.zeros(len(x)), index=x.index)
    return (x - x.mean()) / s

market_metrics["screen_index"] = (
    0.40 * zscore(market_metrics["mean_score"]) +
    0.30 * zscore(market_metrics["pct_score95"]) +
    0.30 * zscore(market_metrics["longest_run_score95"])
)

market_metrics = market_metrics.sort_values("screen_index", ascending=False).reset_index(drop=True)
market_metrics["rank"] = np.arange(1, len(market_metrics) + 1)
```

```python
top_n = 15
top_markets = market_metrics.head(top_n).copy()
top_markets = top_markets.sort_values("screen_index", ascending=True)

fig = px.bar(
    top_markets,
    x="screen_index",
    y="market_id",
    orientation="h",
    hover_data=[
        "mean_score",
        "pct_score95",
        "pct_score95_in_manifold",
        "longest_run_score95",
        "mean_recon"
    ],
    title="Top Markets by Screening Index"
)

fig.update_layout(
    xaxis_title="Screening index",
    yaxis_title="Market"
)

fig.show()
```

```python
fig = px.scatter(
    market_metrics,
    x="mean_recon",
    y="mean_score",
    size="pct_score95_in_manifold",
    hover_name="market_id",
    hover_data=[
        "pct_score95",
        "pct_score95_in_manifold",
        "longest_run_score95",
        "screen_index"
    ],
    title="Market-Level Conduct Score vs Reconstruction Error"
)

# log scale fixes the extreme skew
fig.update_xaxes(type="log")
fig.add_vline(x=recon_tau95, line_dash="dash")
fig.add_hline(y=tau95, line_dash="dash")

fig.update_layout(
    xaxis_title="Mean reconstruction error",
    yaxis_title="Mean conduct score"
)

fig.show()
```

```python
fig = px.scatter(
    market_metrics,
    x="pct_score95_in_manifold",
    y="longest_run_score95",
    hover_name="market_id",
    hover_data=[
        "mean_score",
        "pct_score95",
        "mean_recon",
        "screen_index"
    ],
    title="Persistent Cartel-Like Behaviour Inside the Manifold"
)

fig.update_layout(
    xaxis_title="Share of windows above conduct τ95 and inside manifold",
    yaxis_title="Longest run above conduct τ95"
)

fig.show()
```

## Real martets over time

```python
markets = [
    "ULSP: Pump price (p/litre)",
    "ULSD: Pump price (p/litre)",
    "Nickel"
]

df_plot = score[score["Name"].isin(markets)].copy()
```

```python
fig = px.line(
    df_plot,
    x="Window",
    y="conduct_score_centered",
    color="Name",
    title="Conduct Score Over Time",
    labels={
        "conduct_score_centered": "Centered Conduct Score",
        "date": "Date",
        "Name": "Market"
    }
)

fig.add_hline(
    y=tau95,
    line_dash="dash",
    annotation_text="τ95 threshold",
    annotation_position="top left"
)
fig.update_layout(
    xaxis_title="Rolling time window (18-month window)",
    yaxis_title="Centered conduct score"
)
fig.show()

```

```python
df_plot["smooth_score"] = (
    df_plot.groupby("Name")["conduct_score_centered"]
    .transform(lambda x: x.rolling(5, center=True).mean())
)
```

```python
fig = px.line(
    df_plot,
    x="Window",
    y="smooth_score",
    color="Name",
    title="Conduct Score Over Time",
    labels={
        "conduct_score_centered": "Centered Conduct Score",
        "date": "Date",
        "Name": "Market"
    }
)

fig.add_hline(
    y=tau95,
    line_dash="dash",
    annotation_text="τ95 threshold",
    annotation_position="bottom left"
)
fig.update_layout(
    xaxis_title="Rolling time window (18-month window)",
    yaxis_title="Centered conduct score"
)
fig.show()
```
