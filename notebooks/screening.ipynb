# Evaluation of Screening metrics

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
from pathlib import Path
from scipy.stats import spearmanr, pearsonr

```

```python
#loading data
df = pd.read_parquet("../runs/dgp0/seed_42/baseline/screen/screen_L18.parquet")
df.head()
```

```python
cols = [
    "mean_score",
    "pct_above_tau95",
    "pct_above_tau99",
    "longest_run_tau95",
    "sd_score",
    "max_score",
    "true_pct_cartel_windows",
    "true_pct_pure_cartel_windows",
    "true_mean_share_K",
    "true_longest_cartel_run",
    "true_max_share_K",
    "structural_intensity", 
    "effective_structural_intensity"
]

cor = df[cols].corr()

fig = px.imshow(
    cor,
    text_auto=True,           
    color_continuous_scale='Viridis', 
    title='Correlation Matrix'
)

fig.show()
```

```python
for col in cols:
    spear_cor = coef, p_value = spearmanr(df[col], df["effective_structural_intensity"])
    print(f"Spearman correlation for {col}: {coef:.3f}, p-value: {p_value:.3f}")
```

```python
for col in cols:
    spear_cor = coef, p_value = spearmanr(df[col], df["true_mean_share_K"])
    print(f"Spearman correlation for {col}: {coef:.3f}, p-value: {p_value:.3f}")
```

```python
for col in cols:
    spear_cor = coef, p_value = spearmanr(df[col], df["true_max_share_K"])
    print(f"Spearman correlation for {col}: {coef:.3f}, p-value: {p_value:.3f}")
```

```python
for col in cols:
    spear_cor = coef, p_value = spearmanr(df[col], df["true_longest_cartel_run"])
    print(f"Spearman correlation for {col}: {coef:.3f}, p-value: {p_value:.3f}")
```

```python
fig = px.scatter(
    df,
    x="effective_structural_intensity",
    y="mean_score",
    hover_data=["market_id"],
    title="Market Mean Conduct Score vs Effective Structural Intensity",
    template="plotly_white", 
    trendline="ols", 
    trendline_color_override="orange"
)
fig.show()
```

```python
fig = px.scatter(
    df,
    x="effective_structural_intensity",
    y="pct_above_tau95",
    hover_data=["market_id"],
    title="Share of Suspicious Windows vs Effective Structural intensity ",
    template="plotly_white", 
    trendline="ols",
    trendline_color_override="orange"
)
fig.show()
```

## Structural intensity

```python
r_p, p_p = pearsonr(market_eval["mean_score"], market_eval["effective_structural_intensity"])
r_s, p_s = spearmanr(market_eval["mean_score"], market_eval["effective_structural_intensity"])

print("Pearson :", r_p, p_p)
print("Spearman:", r_s, p_s)
```
