# Evaluation of Screening metrics

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
from pathlib import Path
from scipy.stats import spearmanr

```

```python
#loading data
df = pd.read_parquet("../runs/dgp0/seed_42/baseline/screen/screen_L18.parquet")
df.head()
```

```python
cols = [
    "mean_score",
    "pct_above_tau95",
    "pct_above_tau99",
    "longest_run_tau95",
    "sd_score",
    "max_score",
    "true_pct_cartel_windows",
    "true_pct_pure_cartel_windows",
    "true_mean_share_K",
    "true_longest_cartel_run",
]

cor = df[cols].corr()

fig = px.imshow(
    cor,
    text_auto=True,           
    color_continuous_scale='Viridis', 
    title='Correlation Matrix'
)

fig.show()
```

```python
for col in cols:
    spear_cor = coef, p_value = spearmanr(df[col], df["true_mean_share_K"])
    print(f"Spearman correlation for {col}: {coef:.3f}, p-value: {p_value:.3f}")
```

```python
fig = px.scatter(
    df,
    x="true_mean_share_K",
    y="mean_score",
    hover_data=["market_id"],
    title="Market Mean Conduct Score vs True Mean Cartel Exposure",
    template="plotly_white", 
    trendline="ols", 
    trendline_color_override="orange"
)
fig.show()
```

```python
fig = px.scatter(
    df,
    x="true_mean_share_K",
    y="pct_above_tau95",
    hover_data=["market_id"],
    title="Share of Suspicious Windows vs True Mean Cartel Exposure",
    template="plotly_white", 
    trendline="ols",
    trendline_color_override="orange"
)
fig.show()
```
