# Figures


```python
import pandas as pd
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

## Logit vs AE

```python
df_perf = pd.DataFrame({
    "model":[
        "Logit caseline", 
        "Logit β-only",
        "Logit κ-only",
        "AE β-only",
        "AE κ-only",
        "AE baseline"
    ],
    "AUC":[0.88,0.70,0.69,0.77,0.72,0.88]
})

fig = px.bar(df_perf,x="model",y="AUC")
fig.show()
```

```python
# -------------------------
# Panel A: AUC comparison
# -------------------------
df_auc = pd.DataFrame({
    "setting": ["Baseline", "Beta-only", "Kappa-only"] * 2,
    "model": ["AE"] * 3 + ["Logit"] * 3,
    "value": [0.865, 0.771, 0.728, 0.880, 0.701, 0.693],
})

color_map = {
    "AE": "#1f77b4",     # blue
    "Logit": "#d62728"   # red
}
# -------------------------
# Panel B: Calm FPR
# -------------------------
df_fpr = pd.DataFrame({
    "threshold": ["τ95", "τ99"] * 2,
    "model": ["AE"] * 2 + ["Logit"] * 2,
    "value": [0.131, 0.055, 0.507, 0.296],
})

fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=("Cross-World ROC-AUC", "Calm Fundamentals False Positive Rate")
)

# AUC traces
for model in ["AE", "Logit"]:
    d = df_auc[df_auc["model"] == model]
    fig.add_trace(
        go.Bar(
            x=d["setting"],
            y=d["value"],
            name=model,
             marker_color=color_map[model]
        ),
        row=1, col=1
    )

# FPR traces
for model in ["AE", "Logit"]:
    d = df_fpr[df_fpr["model"] == model]
    fig.add_trace(
        go.Bar(
            x=d["threshold"],
            y=d["value"],
            name=model,
            marker_color=color_map[model],
            showlegend=False
        ),
        row=1, col=2
    )

fig.update_layout(
    title="Autoencoder vs Logistic Regression: Generalization and Robustness",
    template="plotly_white",
    barmode="group",
    width=1100,
    height=500
)

fig.update_yaxes(title_text="ROC-AUC", row=1, col=1, range=[0, 1])
fig.update_yaxes(title_text="False Positive Rate", row=1, col=2, range=[0, 0.6])

fig.show()
```
